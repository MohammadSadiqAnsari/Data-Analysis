```python
#Extracting the records where a team won batting first
batting_first = ipl[ipl["win_by_runs"]!= 0]
batting_first
```

```python
#Making a bar-Plot for top 5 teams who bat first and won
plt.figure(figsize = (12,7))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first["winner"].value_counts()[0:5]),color = ["blue","black","orange","green","pink"])
plt.title("Bar Plot")
plt.xlabel("Teams")
plt.ylabel("No of winnings")
plt.show()
```

```python
#Making a histogram of the teams who bat first and won the match
runs = batting_first['win_by_runs'][:]
plt.figure(figsize=(15,10))
plt.hist(runs,color=["pink"],bins=30)
plt.title("Histogram for win by runs")
plt.xlabel("runs")
plt.ylabel("frequency")
plt.show()
```

```python
#Making a bar-plot for top 3 teams with most wins after fielding first
plt.figure(figsize=(7,7))
plt.bar(list(field_first['winner'].value_counts()[0:3].keys()),list(field_first['winner'].value_counts()[0:3]),color=["blue","yellow","orange"])
plt.title("Bar Plot")
plt.xlabel("Teams")
plt.ylabel("No of winnings")
plt.show()
```

```python
#Making a pie chart for teams won after batting first
plt.figure(figsize=(10,10))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title("Won after batting first")
plt.show()
```

```python
#Making a pie chart for teams won after fielding first
plt.figure(figsize=(10,10))
plt.pie(list(field_first['winner'].value_counts()),labels=list(field_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title("Won after fielding first")
plt.show()
```
